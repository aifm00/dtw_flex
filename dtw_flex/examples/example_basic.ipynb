```python
#%pip install ipympl
%matplotlib ipympl
from dtw_flex import Dtw
import numpy as np
```

```python
nref = 3 #length of reference array
nquery = 12 #length of query array
dtw_obj = Dtw(nref, nquery) #init Dtw class
```

```python
dtw_obj.flag = np.array([dtw_obj.nref // 2]) #highlight middle point in the result. This will be visual in red (possible to change in plot below)
dtw_obj.ref[dtw_obj.nref // 2] = 1 #set middle point value for reference array
dtw_obj.rep_ref[dtw_obj.nref // 2] = 1 #set allowable repetition for middle point in reference array
dtw_obj.ref_plotset(hide_endpoints=True) #create interactive reference plot 
```

```python
dtw_obj.query = np.random.rand(nquery) #create query array
dtw_obj.query[6:8] = dtw_obj.query[6:8] + 2 #add 2 peak values
dtw_obj.query_plotset() #create interactive query plot
```

```python
dtw_obj.run_dtw(plt_fit = True, plt_match = True) #run example
#first plot shows the match between the altered (offset + amplitude) reference array and the query array. "reps" shows the poins which are repeated.
#the second plot shows the flagged (red) points of the reference array matching the query array
```
